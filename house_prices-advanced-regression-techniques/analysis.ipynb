# house-price 기초분석

### MSSubClass : 주택 유형 (범주)
* 20,30,40,45,50,60,70,75,80,85,90,120,150,160,180,190
### MSZoning : 주택 용도(범주)
* RL, RM, C (all), FV, RH
* C (all) : ALL로 변환 시키기
### LotFrontage : 건물과 도로 사이 거리(수치)
### LotArea : 도로 크기(수치)
### Street : 도로 타입(범주)
* Pave, Grvl
* pave : 포장도로, Grvl : 자갈
### Alley : 마당 길? 집 드가는 좁은 길(범주)
* Pave, Grvl
* nan 값 O
### LotShape : 도로 모양(범주)
* Reg, IR1, IR2, IR3
* Reg : 균형적, IR1 : 약간 불균형, IR2 : 많이 불균형, IR3 : 완전 불균형
* 선호도 표시하듯이 바꿀 수 있을 거 같음
### LandContour : 토지 평탄도(범주)
* Lvl, Bnk, Low, HLS
* Lvl : 매우 평탄, Bnk : 도로와 건물 사이의 높은 경사, Low : 하강, HLS : 좌우로 심한 경사
### Utilities : 유용성(범주)
* AllPub, NoSeWa
* (전기, 가스, 물, 정화조) 다 있으면 AllPub, (전기,가스) 있으면 NoSeWa
* NoSeWa 값 1개라 지워도 될듯
### LotConfig : 도로 분류(범주)
* Inside, FR2, Corner, CulDSac, FR3
* FR2 : 두 방향 길, FR3 : 세 방향 길, CulDSac : 막다른 길
### LandSlope: 토지 경사(범주)
* Gtl, Mod, Sev
* Gtl : 경사 없음, Mod : 경사 조금 있음, Sev : 경사 심함
* 선호도처럼 바꿔도 될듯
### Neighborhood : 에임스 지역 내 분류(범주)
* CollgCr, Veenker, Crawfor, NoRidge, Mitchel ,Somerst, NWAmes,
  OldTown, BrkSide, Sawyer, NridgHt, NAmes, SawyerW, IDOTRR,
  MeadowV, Edwards, Timber, Gilbert, StoneBr, 'ClearCr, NPkVill,
  Blmngtn, BrDale, SWISU, Blueste
* 집값과 높은 연관 있을 거 같음, 지역 내에서 비싼 동네, 싼 동네로 컬럼 새로 하나 만들면 좋을듯
### Condition1 : 주변  환경(1개만 존재하는 것)
* Norm, Feedr, PosN, Artery, RRAe, RRNn, RRAn, PosA, RRNe
* Norm : 적당하다, Feedr : 지선도로(마을버스가 지선버스임), PosN : 공원, Artery : 간선도로(파란버스가 간선버스), RRAe : 동부-서부 잇는 철도, RRNn : 남부-북부 잇는 철도 200피트 이내, RRAn : 남부-북부 잇는 철도, PosA : 공원과 인접, RRNe : 동부-서부 철도 200피트 이내
* 인접과 200피트 이내 차이를 모르겠음 둘이 합치던가 제거 하든가 해야함.
* 주변 교통상황 매우 중요할 것으로 보이는데 애매함 Norm, Feeder, PosN, Artery 정도로 추려야 할듯
### Condition2 : 주변  환경(2개 이상 갖고 있는 것)
* 위와 같음
* Norm 이외에 10개 정도임. 나머지 중요할듯.
### BldgType : 주거 타입(범주)
* 1Fam, 2fmCon, Duplex, TwnhsE, Twnhs
* 1Fam : 1가구, 2fmCon : 2가구, Duplex : 땅콩집(하나의 부지에 두개의 집 값이 쌈), TwnhsE : 타운하우스 끝자락 집(Twnhsl 보다 비쌈), TwnhsI : 타운하우스 끝자락 외의 집
* 땅콩집 저렴, TwnhsE > Twnhsl
### HouseStyle : 집 스타일(범주)
* 2Story, 1Story, 1.5Fin, 1.5Unf, SFoyer, SLvl, 2.5Unf, 2.5Fin
* stroy : 층수, 1.5Fin : 1.5층 완공, 1.5Unf : 1.5층 미완공, SFoyer : 반층계 위 집, SLvl : 반층계 아래 집
### OverallQual : 전체적인 품질(범주)
* 1~10
### OverallCond : 전체적인 상태(범주)
* 1~10
### YearBuilt : 건축년도(수치)
* 1870 년대 ~ 2000 년대
* OverallQual, OverallCond 하고 같이 비교하면 좋을듯
### YearRemodAdd : 리모델링년도(수치)
* 리모델링 x = 건축년도 그대로
* 건축년도 제외하고 이거 보고하는게 좋을듯
### RoofStyle : 지붕스타일(범주)
* Gable, Hip, Gambrel, Mansard, Flat, Shed
* 큰 영향 없을 거 같음.
### RoofMatl : 지붕 재료(범주)
* x
### Exterior1st : 외부 벽 자재(범주)
* 재료 이름임.
### Exterior2nd : 외부 벽 자재(2개 이상)
* 위외 같음
### MasVnrType, MasVnrArea 제거
### ExterQual : 외장재 품질(범주)
* Ex, Gd, TA, Fa, Po 순으로
* over 컬럼들이랑 비교해봐도 좋을듯
### ExterCond : 외장재 상태(범주)
* 위와 같음
### Foundation: 건물 기반 자재(범주)
* PConc, CBlock, BrkTil, Wood, Slab, Stone
* PConc : 콘크리트, CBlock : Cinder Block
* 콘크리트와 신더블럭 두개가 600개 압도적으로 많음.
### BsmtQual : 지하실 높이(범주)
* Ex, Gd, TA, Fa, Po, NA

### BsmtCond : 지하실 상태(범주)
* 위와 같음
### BsmtExposure : 지하실 외부 노출 정도
* X
### BsmtFinType1, BsmtFinSF1, BsmtFinType2, BsmtFinSF2, BsmtUnfSF, TotalBsmtSF 제거
### Heating : 보일러 종류(범주)

### Electrical : 전기 타입(범주)
